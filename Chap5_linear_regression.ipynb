```python
import torch
import torch.nn as nn
from matplotlib import pyplot as plt
```

```python
x=torch.FloatTensor(range(5)).unsqueeze(1) #2차원 데이터로 표현
y=2*x+torch.rand(5,1)

class LinearRegressor(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc=nn.Linear(1,1,bias=True)

  def forward(self,x):
    y=self.fc(x)
    return y
```

```python
#Define model
model=LinearRegressor()
learning_rate=1e-3
criterion=nn.MSELoss()
optimizer=torch.optim.SGD(model.parameters(),lr=learning_rate) #최적회 방법에 모델 파라미터 넣어줌

loss_stack=[]
for epoch in range(1001):
  optimizer.zero_grad()

  y_hat=model(x)
  loss=criterion(y_hat,y)

  loss.backward()
  optimizer.step()
  loss_stack.append(loss.item())

  if epoch % 100==0:
    print(f'Epoch {epoch}:{loss.item()}')
```
